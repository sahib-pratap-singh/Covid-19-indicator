```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\sahib pratap\\Downloads\\AgeGroupDetails.csv')
df.drop(df.tail(1).index,inplace=True)
df['Percentage']=df['Percentage'].str.strip('%')
```

```python
df['Percentage']=df['Percentage'].map(float)
```

# Agegroup percentage of cases in India

```python
plt.figure(figsize=(10,5))
plt.bar(df['AgeGroup'],df['Percentage'],color='violet')
plt.title('Agegroup percentage of cases')  
plt.xlabel('Age group')
plt.ylabel('Percentage')
```

```python
sns.stripplot(x=df['AgeGroup'], y=df['Percentage'], 
              data=df, dodge=True, alpha=.25, zorder=1)
sns.pointplot(x=df['AgeGroup'], y=df['Percentage'],
              data=df, dodge=.532, join=False, palette="dark",
              markers="d", scale=.75, ci=None)
```

```python
df = pd.read_csv("C:\\Users\\sahib pratap\\Downloads\\Hospitals_and_Beds_statewise.csv")
df.style.set_properties(**{'background-color': 'green',
                            'color': 'gold',
                            'border-color': 'white'})
```
